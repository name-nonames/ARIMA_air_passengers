```python
log_data = np.log(df_data)
plt.plot(log_data)
```

```python
movingavg = log_data.rolling(window =12).mean()
#movingstd = log_data.rolling(window =12).std()
log_data_I = log_data - log_data.shift()
log_data_I.dropna(inplace=True)
plt.plot(log_data_I)
```

```python
log_data_I_test = adfuller(log_data_I["#Passengers"])
log_data_I_test
```
